```python
import numpy as np
import matplotlib.pyplot as plt
```

### DFT of sin(5t) 

```python
x = np.linspace(0,2*np.pi, 129)
x = x[0:-1]
y = np.sin(5*x)
Y = np.fft.fft(y)
Y = np.fft.fftshift(Y)/128
w = np.linspace(-64,64,129)
w = w[0:-1]

plt.figure()
plt.subplot(2,1,1)
plt.plot(w, abs(Y),lw=2)
plt.ylabel(r"$|Y|\rightarrow$")
plt.xlim([-10,10])
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w,np.angle(Y),'ro',lw=2)
ii=np.where(abs(Y)>1e-3)
plt.plot(w[ii],np.angle(Y[ii]),'go',lw=2)
plt.xlim([-10,10])
plt.ylabel(r"Phase of $Y\rightarrow$")
plt.xlabel(r"$k\rightarrow$")
plt.grid(True)
plt.savefig("fig9-2.png")
plt.show()
```

### DFT of (1+0.1cos(x))cos(10x)

```python
x = np.linspace(0,2*np.pi, 129)
x = x[0:-1]
y = (1 + 0.1*np.cos(x))*np.cos(10*x)
Y = np.fft.fft(y)
Y = np.fft.fftshift(Y)/128
w = np.linspace(-64,64,129)
w = w[0:-1]
plt.figure()

plt.subplot(2,1,1)
plt.plot(w, abs(Y),lw=2)
plt.ylabel(r"$|Y|\rightarrow$")
plt.xlim([-15,15])
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(w,np.angle(Y),'ro',lw=2)
ii=np.where(abs(Y)>1e-3)
plt.plot(w[ii],np.angle(Y[ii]),'go',lw=2)
plt.xlim([-15,15])
plt.ylabel(r"Phase of $Y\rightarrow$")
plt.xlabel(r"$k\rightarrow$")
plt.grid(True)
plt.savefig("fig9-2.png")
plt.show()
```
